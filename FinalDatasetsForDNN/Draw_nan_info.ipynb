```python
import pandas as pd 
import missingno as msno 
import sys
import os
%matplotlib inline
def get_nan_rate(df):
    msno.matrix(df)
    msno.heatmap(df)
#     msno.bar(data)
    l = []
    size = len(df)
    print(size)
    for col in df.columns:
        #print(col)
        nan_num = df[[col]].isna().sum()
        rate = nan_num/size
        #print(nan_num)
        l.append(rate.values[0]*100)
        #print(rate*100)
    s = pd.Series(l,index= df.columns)
    print(s)
print(os.getcwd())
print('air pollution')
aq_data = pd.read_csv("17_18aq.csv")
i=0
for id in  aq_data['station_id'].unique():
    print(i)
    print(id)
    i+=1
    print(aq_data[aq_data['station_id']==id].shape)
get_nan_rate(aq_data)
```
