```python
import pandas as pd 
import missingno as msno 
import sys
import os
%matplotlib inline
def get_nan_rate(df):
    msno.matrix(df)
    msno.heatmap(df)
#     msno.bar(data)
    l = []
    size = len(df)
    print(size)
    for col in df.columns:
        #print(col)
        nan_num = df[[col]].isna().sum()
        rate = nan_num/size
        #print(nan_num)
        l.append(rate.values[0]*100)
        #print(rate*100)
    s = pd.Series(l,index= df.columns)
    print(s)
```

## missingno.heatmap函数
### Nullity correlation ranges from -1 (if one variable appears the other definitely does not) to 0 (variables appearing or not appearing have no effect on one another) to 1 (if one variable appears the other definitely also does).

# 北京2014-2015年数据缺失情况

```python
print(os.getcwd())
print('air pollution')
aq_data = pd.read_csv("beijing20142015/beijing14_15_aq.csv")
get_nan_rate(aq_data)
print('meo')
meo_data = pd.read_csv("beijing20142015/beijing14_15_meo.csv")
get_nan_rate(meo_data)
```

```python
print('air pollution')
aq_data = pd.read_csv("beijing20172018/beijing17_18_aq.csv")
print(aq_data.shape)
for id in  aq_data['station_id'].unique():
    print(aq_data[aq_data['station_id']==id].shape)
get_nan_rate(aq_data)
print('meo')
meo_data = pd.read_csv("beijing20172018/beijing17_18_meo.csv")
get_nan_rate(meo_data)
```
